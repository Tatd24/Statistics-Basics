**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

Ans: Statistics is broadly divided into two main branches — Descriptive Statistics and Inferential Statistics.
They serve different purposes in data analysis.

1. Descriptive Statistics

Definition:
Descriptive statistics are used to summarize, organize, and describe the main features of a dataset.
It deals only with the data we have (no generalization beyond it).

Purpose:
To provide a clear overview of the data using numerical and graphical methods.

Common Techniques:

*   Measures of Central Tendency: Mean, Median, Mode
*   Measures of Dispersion: Range, Variance, Standard Deviation
*   Data Visualization: Bar charts, Pie charts, Histograms, Box plots

Inferential Statistics
Definition:
Inferential statistics use sample data to make conclusions, predictions, or inferences about a larger population.
Purpose:
To go beyond the available data and generalize results using probability theory.
Common Techniques:
*	Hypothesis Testing (e.g., t-test, ANOVA)
*	Confidence Intervals
*	Regression Analysis
*	Correlation and Prediction
Example:
You collect the average height from a sample of 50 students to estimate the average height of all students in the school.
You might say:
“We are 95% confident that the average height of all students is between 162 cm and 168 cm.”
Here, you’re using sample data to infer about the population.

What is sampling in statistics? Explain the differences between random and stratified sampling.**

Ans- Sampling is the process of selecting a small group (called a sample) from a larger group (called the population) in order to study the population’s characteristics.
Because studying an entire population is often time-consuming and expensive, researchers collect data from a sample and use it to make inferences about the whole population.

Types of Sampling
1. Random Sampling
•	In random sampling, every individual in the population has an equal chance of being selected.
•	It helps to avoid bias and ensures that the sample represents the population fairly.
•	Example: If you have 1,000 students, and you randomly pick 100 using a random number generator, that’s random sampling.
Advantages:
•	Simple to use.
•	Reduces selection bias.
Disadvantages:
•	Might not perfectly represent subgroups (e.g., gender, age, income levels) if they vary widely.

2. Stratified Sampling
•	In stratified sampling, the population is divided into subgroups (strata) based on specific characteristics (like age, gender, or income).
•	Then, random samples are taken from each stratum proportionally.
•	Example: If a school has 60% boys and 40% girls, you could select 60 boys and 40 girls randomly from each group to form a 100-student sample.
Advantages:
•	Ensures all key subgroups are represented.
•	Produces more accurate and reliable results than simple random sampling, especially when population groups differ greatly.
Disadvantages:
•	More complex and time-consuming to organize.
•	Requires prior knowledge about the population’s composition.

Define mean, median, and mode. Explain why these measures of central
tendency are important.**

Ans- Measures of central tendency are statistical values that describe the center or typical value of a dataset.
The three most common measures are Mean, Median, and Mode.

1. Mean (Average)
•	The mean is the sum of all data values divided by the total number of values.
•	Formula:
Mean=Sum of all observationsNumber of observations\text{Mean} = \frac{\text{Sum of all observations}}{\text{Number of observations}}Mean=Number of observationsSum of all observations
•	Example:
For data: 5, 10, 15,
Mean=5+10+153=10\text{Mean} = \frac{5 + 10 + 15}{3} = 10Mean=35+10+15=10
Note: The mean is sensitive to extreme values (outliers).

2. Median
•	The median is the middle value when all data points are arranged in order.
•	If there’s an even number of data points, the median is the average of the two middle values.
Example:
Data: 5, 10, 15 → Median = 10
Data: 4, 8, 10, 12 → Median = (8 + 10) / 2 = 9
Note: The median is not affected by outliers, making it useful for skewed data.

3. Mode
•	The mode is the value that appears most frequently in the dataset.
Example:
Data: 3, 5, 5, 7, 9 → Mode = 5
If no number repeats, there is no mode.
If multiple numbers repeat equally often, the data is bimodal or multimodal.

Importance of These Measures
Measure	Key Use	Strength
Mean	Represents the average value of data	Useful for numerical, balanced data
Median	Shows the central value of ordered data	Best for skewed or unevenly distributed data
Mode	Identifies the most common value	Useful for categorical or frequency data



**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

 Ans- Both skewness and kurtosis are measures that describe the shape of a data distribution beyond just its center (mean) and spread (variance).

1. Skewness
Definition:
Skewness measures the asymmetry of a data distribution around its mean.
•	If a distribution is symmetrical, its left and right sides are mirror images, and skewness = 0.
•	If it is not symmetrical, it is said to be skewed.
Types of Skewness:
Type	Description	Shape
Positive Skew (Right Skew)	Tail is longer on the right side of the mean. Most data values are concentrated on the left.	Mean > Median > Mode
Negative Skew (Left Skew)	Tail is longer on the left side of the mean. Most data values are concentrated on the right.	Mean < Median < Mode
Zero Skew (Symmetric)	Data is evenly distributed around the mean.	Mean = Median = Mode

✅ What a Positive Skew Implies
A positive skew (right-skewed distribution) means:
•	Most of the data values are small, clustered on the left side.
•	A few very large values (outliers) stretch the tail to the right.
•	The mean is pulled higher than the median.
Example:
Income data is often positively skewed — most people earn moderate incomes, while a few earn extremely high incomes.

2. Kurtosis
Definition:
Kurtosis measures the “tailedness” or peakedness of a data distribution — that is, how heavy or light the tails are compared to a normal distribution.
Types of Kurtosis:
Type	Description	Shape
Mesokurtic (K = 3)	Normal distribution; moderate tails	Standard bell shape
Leptokurtic (K > 3)	Heavy tails; sharp peak; more outliers	More extreme values
Platykurtic (K < 3)	Light tails; flatter peak	Fewer outliers

Summary Table
Measure	What It Describes	Positive Value Means
Skewness	Asymmetry of data	Right tail longer (mean > median)
Kurtosis	Tailedness of data	Sharper peak, more outliers



**Question 5:Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.)**

```python
# Given list of numbers
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculating Mean, Median, and Mode
mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)

# Displaying the results
print("List of Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

**Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] (Include your Python code) **

```python
import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=False)

# Extract covariance value
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Display results
print("List X:", list_x)
print("List Y:", list_y)
print("Covariance:", round(covariance, 2))
print("Correlation Coefficient:", round(correlation, 2))
```

**Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of Given Data")
plt.xlabel("Values")
plt.show()

# Identify outliers using IQR method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

**Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]**

Ans-  import statistics as stats
Concept Explanation
As a data analyst, to explore whether advertising spend affects daily sales, you can use covariance and correlation:
1. Covariance
•	Covariance measures how two variables vary together.
•	If the covariance is positive, it means as advertising spend increases, daily sales also tend to increase.
•	A negative covariance would mean that higher spending is associated with lower sales.
•	However, covariance does not tell the strength of the relationship (it depends on the scale of data).
2. Correlation
•	Correlation is a standardized form of covariance that measures both direction and strength of a linear relationship.
•	It is dimensionless and ranges between –1 and +1:
o	+1 → perfect positive relationship
o	–1 → perfect negative relationship
o	0 → no linear relationship
Thus, correlation is more useful for comparing relationships because it’s scale-independent.



```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Display results
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Covariance:", round(covariance, 2))
print("Correlation Coefficient:", round(correlation, 3))
```

**Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

Ans- To understand the distribution of customer satisfaction scores (on a 1–10 scale), we use summary statistics and visualizations that describe the central tendency, spread, and shape of the data.

1. Summary Statistics to Use
Measure	Purpose
Mean	Shows the average satisfaction level.
Median	Indicates the middle satisfaction score (less affected by outliers).
Mode	Reveals the most common rating.
Standard Deviation (SD)	Measures how spread out the scores are — a low SD means consistent opinions, while a high SD means varied opinions.
Range	Shows the gap between the lowest and highest ratings.
2. Visualization to Use
Visualization	Purpose
Histogram	Shows how often each score appears, helping to visualize the shape (e.g., symmetric, skewed).
Boxplot (optional)	Displays median, quartiles, and potential outliers.

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

# Given survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
mode_score = stats.mode(survey_scores)
std_dev = np.std(survey_scores)
score_range = max(survey_scores) - min(survey_scores)

# Display summary
print("Customer Satisfaction Survey Summary:")
print("Mean:", round(mean_score, 2))
print("Median:", median_score)
print("Mode:", mode_score)
print("Standard Deviation:", round(std_dev, 2))
print("Range:", score_range)

# Create a histogram
plt.hist(survey_scores, bins=6, color='skyblue', edgecolor='black')
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
